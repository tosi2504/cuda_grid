### Matmul Performance:
Comparing the mrhs-matmul performance of several approaches.

#### Plots:
From left to right:  
- straight-forward SIMT approach (target: "mrhs_lanes")  
- cuBLAS (including the necessary copy for layout change) (target: "mrhs_blas")
- sharedmem: explicitly controlling the shared memory (cache) of the GPU (target: "mrhs_sharedmem")

The `srhs_bw(GBs)` is the bandwidth calculated as if the operation was single right hand sided, i.e., for every matmul operation the whole matrix was loaded again. 
srhs_bw = gridpoints * numRHS * (N**2 + 2N)*sizeof(T)/delta_t  
This is a good metric to measure improvement over a single RHS approach.

The `mrhs_bw(GBs)` is the bandwidth calculated properly. I.e. mrhs_bw = gridpoints * (N**2 + 2 * numRHS * N) * sizeof(T) / delta_t

The `time(us)` is simply the average time over 100 repetitions.

#### What to do:
Play around with the entries in the `selectors` dict and the `measurement` variable (the available values are printed :))

```python
fig, [ax_lanes, ax_blas, ax_shmem] = plt.subplots(nrows=1, ncols=3, figsize=(5,4), sharey=True)
print("available grids           :", pyBenchmark.grids)
print("available mat sizes       :", pyBenchmark.Ns)
print("available number of rhs's :", pyBenchmark.numRHSs)
print("available bench outputs   :", str(['time(us)', 'srhs_bw(GBs)', 'mrhs_bw(GBs)']))

# play around with this line:
selectors = {'grid': '(8, 8, 8, 8)', 'N': 128, 'numRHS': 60}
measurement = 'srhs_bw(GBs)'

fig.suptitle(str(selectors))
plot_data(df=df, ax=ax_lanes,   xlabel="type T", valuelabel=measurement, selectors=selectors | {'target':'mrhs_lanes'}    , title='lanes')
plot_data(df=df, ax=ax_blas,    xlabel="type T", valuelabel=measurement, selectors=selectors | {'target':'mrhs_blas'}     , title='blas')
plot_data(df=df, ax=ax_shmem,   xlabel="type T", valuelabel=measurement, selectors=selectors | {'target':'mrhs_sharedmem'}, title='shmem')

fig.tight_layout()
plt.show()
```

### Stencil performance:
Compare the performance of two approaches to simple stencil operations.
Simple stencil in this case means that a vector is propagated in one direction along a link.
Any stencil or stapel could be build from such an operation.
We only compare the blas approach with the SIMT(lanes) approach. 
In principal huge improvements could be done by applying caching of the rhs vectors as soon as more than one direction is included in the stencil.
This might be interesting for future work.

#### Plots:
- left: SIMT approach (lanes)
- right: blas approach 

Available metrics are the same as for mrhs-matmul (see above).

#### Notes:

##### Strided vs non strided GEMV:
The stencil-blas routine uses the non-strided batched version of GEMV, while the matmul-blas routine uses the strided batched version.
This is most likely the main reason why they are different in performance, though this would need to be confirmed, as there is also some additional overhead to the stencil routine from processing geometry-specific stuff.
In principal one does not need the non-strided version for stencils, but could implement the stencil using only strided GEMV calls ... it just would need many seperate calls.
Could be interesting to see whether that would improve performance.

Also, since all blas routines require a layout change, one could perform this change in such a way, that the GEMV call simplifies.
The problem with that is that subsequent stencil calls with different directions would each require a seperate layout change. 

##### Implicit copy and layout change:
I am pretty sure that the layout change is far from optimal (this can be seen in the next section). 
This ofc impacts the performance of the blas calls. 
Since the layout change duration is still short relative to the overall duration of the stencil operation, I have not put further effort into this.

```python
fig, [ax_lanes, ax_blas] = plt.subplots(nrows=1, ncols=2, figsize=(5,4), sharey=True)
print("available grids           :", pyBenchmark.grids)
print("available mat sizes       :", pyBenchmark.Ns)
print("available number of rhs's :", pyBenchmark.numRHSs)
print("available bench outputs   :", str(['time(us)', 'srhs_bw(GBs)', 'mrhs_bw(GBs)']))

# play around with this line:
selectors = {'grid': '(8, 8, 8, 8)', 'N': 128, 'numRHS': 60}
measurement = 'srhs_bw(GBs)'

fig.suptitle(str(selectors))
plot_data(df=df, ax=ax_lanes,   xlabel="type T", valuelabel=measurement, selectors=selectors | {'target':'stencil_lanes'}    , title='lanes')
plot_data(df=df, ax=ax_blas,    xlabel="type T", valuelabel=measurement, selectors=selectors | {'target':'stencil_blas'}     , title='blas')

fig.tight_layout()
plt.show()
```

### Layout change and matmul vs stencil
Prior to any blas calls the data layout needs to be 'untagled'.
I.e. spacial indices must be seperated from tensor indices. 
Later it needs to be entangled again.
To this end, a copy function has been implemented.
In the following this copy function is presented.

#### Plots:
- left: copy operation
- middle: blas operation for stencil
- right: blas operation for matmul

#### Notes:
This really shows how 'bad' the stencil is compared to matmul when it comes to blas calls.
I am sure there is some way to improve this discrepancy.
Also for the matmul operation, the layout change really does take a significant amount of time.

```python
fig, [ax_nomalloc, ax_stencil, ax_matmul] = plt.subplots(nrows=1, ncols=3, figsize=(5,4), sharey=True)
print("available grids           :", pyBenchmark.grids)
print("available mat sizes       :", pyBenchmark.Ns)
print("available number of rhs's :", pyBenchmark.numRHSs)
print("available bench outputs   :", str(['time(us)', 'srhs_bw(GBs)', 'mrhs_bw(GBs)']))

# play around with this line:
selectors = {'grid': '(16, 16, 16, 16)', 'N': 128, 'numRHS': 60}
measurement = 'time(us)'

fig.suptitle(str(selectors))
plot_data(df=df, ax=ax_nomalloc , xlabel="type T", valuelabel=measurement, selectors=selectors | {'target':'copy_noMalloc'} , title='copy')
plot_data(df=df, ax=ax_stencil  , xlabel="type T", valuelabel=measurement, selectors=selectors | {'target':'stencil_blas'}  , title='stencil')
plot_data(df=df, ax=ax_matmul   , xlabel="type T", valuelabel=measurement, selectors=selectors | {'target':'mrhs_blas'}     , title='matmul')

fig.tight_layout()
plt.show()
```
